```python
from scipy.stats import f_oneway
```

<p>Requirement for ONE WAY ANNOVA</p>
<p>Filter Columns and its Data</p>

```python
birds[birds['sex']== 'male']['body_mass_g']
```

```python
birds[birds['sex']== 'female']['body_mass_g']
```

```python
birds[birds['sex']== 'male']['body_mass_g']

f_stats, p_value = f_oneway(birds[birds['sex']== 'male']['body_mass_g'],birds[birds['sex']== 'female']['body_mass_g'])

print(p_value)
```

<h1>Insights No.16</h1>
<p>There's a correlation between sex & body mass</p>

```python
birds[birds['sex']== 'male']['body_mass_g']

f_stats, p_value = f_oneway(birds[birds['sex']== 'male']['body_mass_g'],
                            birds[birds['sex']== 'female']['body_mass_g'])

print(p_value)
```
