<h1>Data Analytics</h1>

```python
sns.histplot(df['Total Household Income'], bins=50, kde=True)
plt.title("Distribution of Total Household Income")
plt.gcf().set_size_inches(13, 8)
plt.show()
```

<h1>Insight No. 1</h1>
<h2>Total household income is highly right-skewed, indicating that most households earn lower incomes while a few earn very high incomes.</h2>

```python
sns.boxplot(data=df, x='Region', y='Total Household Income')
plt.xticks(rotation=90)
plt.title("Household Income by Region")
plt.gcf().set_size_inches(13, 8)
plt.show()
```

<h1>Insight No. 2</h1>
<h2>Certain regions show significantly higher income levels, such as NCR, while others display more modest distributions.</h2>

```python
sns.histplot(df['Total Food Expenditure'], bins=50, kde=True)
plt.title("Distribution of Total Food Expenditure")
plt.gcf().set_size_inches(13, 8)
plt.show()
```

<h1>Insight No. 3</h1>
<h2>Food expenditure also shows a right-skewed distribution, with the majority spending on the lower side and a few spending much more.</h2>

```python
sns.scatterplot(x='Total Household Income', y='Total Food Expenditure', data=df)
plt.title("Food Expenditure vs Household Income")
plt.gcf().set_size_inches(13, 8)
plt.show()
```

<h1>Insight No. 4</h1>
<h2>A weak positive trend suggests that households with higher incomes tend to spend more on food, although variability is high.</h2>

```python
sns.countplot(data=df, y='Main Source of Income', order=df['Main Source of Income'].value_counts().index)
plt.title("Main Source of Income Distribution")
plt.gcf().set_size_inches(13, 8)
plt.show()
```

<h1>Insight No. 5</h1>
<h2>Most households rely on wages/salaries, followed by entrepreneurial activities, with few relying on property or pension.</h2>

```python
sns.countplot(x='Agricultural Household indicator', data=df)
plt.title("Agricultural vs Non-Agricultural Households")
plt.gcf().set_size_inches(13, 8)
plt.show()
```

<h1>Insight No. 6</h1>
<h2>The majority of households are non-agricultural, indicating urban or non-farming employment is more common.</h2>

```python
sns.scatterplot(x='Meat Expenditure', y='Total Rice Expenditure', data=df)
plt.title("Meat vs Rice Expenditure")
plt.gcf().set_size_inches(13, 8)
plt.show()
```

<h1>Insight No. 7</h1>
<h2>There is a moderate positive relationship between meat and rice expenditure, implying dietary patterns may scale together.</h2>

<h1>GENERAL CONCLUSION</h1>
<P>After analyzing the data, we found that how much a household spends on food depends on their income, the region they live in, and their type of work. Families with higher income usually spend more on food, but not always by a large amount. Rice is one of the biggest food expenses for almost every family. People living in different regions also spend differently, mostly because of the cost of living in those areas. Families who farm or grow their own food spend differently than others — they often spend less on things like meat or processed food. These findings can help businesses or local governments make better decisions, like how to price food or where to give more support.</P>
